#  According to this categorical dataset we are implementing classication problem random forest , logistic regression and decession tress in the end we will pic best accuracy model

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

X=df[['src_nm', 'src_id', 'cas_cas_no', 'cas_pt','cas_source','cas_match','umls_pt','umls_cui','ind_raw_value','ind_raw_target','ind_raw_match','ind_umls_entry_term_match']]  # Features
y=df['drug_raw_name']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
X_train
```

```python
y_train
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
true_positive = conf_mat[0][0]
false_positive = conf_mat[0][1]
false_negative = conf_mat[1][0]
true_negative = conf_mat[1][1]
```

```python
# Breaking down the formula for Accuracy
Accuracy = (true_positive + true_negative) / (true_positive +false_positive + false_negative + true_negative)
Accuracy
```

```python
# Precison
Precision = true_positive/(true_positive+false_positive)
Precision
```

```python
# Recall
Recall = true_positive/(true_positive+false_negative)
Recall
```

```python
# F1 Score
F1_Score = 2*(Recall * Precision) / (Recall + Precision)
F1_Score
```
