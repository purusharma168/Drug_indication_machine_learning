```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

X=df[['src_nm', 'src_id', 'cas_cas_no', 'cas_pt','cas_source','cas_match','umls_pt','umls_cui','ind_raw_value','ind_raw_target','ind_raw_match','ind_umls_entry_term_match']]  # Features
y=df['drug_raw_name']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # 80% training and 30% test
```

```python
X_train
```

```python
y_test
```

```python
from sklearn.ensemble import BaggingClassifier
bag = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=200,
    max_samples=0.25,
    bootstrap=True,
    random_state=42
)
```

```python
bag.fit(X_train,y_train)
```

```python
y_pred = bag.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
bag.estimators_samples_[0].shape
```

```python
bag.estimators_features_[0].shape
```

# Bagging using SVM

```python
from sklearn.svm import SVC
bag = BaggingClassifier(
    base_estimator=SVC(),
    n_estimators=200,
    max_samples=0.25,
    bootstrap=True,
    random_state=42
)
```

```python
bag.fit(X_train,y_train)
y_pred = bag.predict(X_test)
print("Bagging using SVM",accuracy_score(y_test,y_pred))
```

# Pasting

```python
bag = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=200,
    max_samples=0.25,
    bootstrap=False,
    random_state=42,
    verbose = 1,
    n_jobs=-1
)
```

```python
bag.fit(X_train,y_train)
y_pred = bag.predict(X_test)
print("Pasting classifier",accuracy_score(y_test,y_pred))
```

# Random Subspaces

```python
bag = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=200,
    max_samples=1.0,
    bootstrap=False,
    max_features=0.5,
    bootstrap_features=True,
    random_state=42
)
```

```python
bag.fit(X_train,y_train)
y_pred = bag.predict(X_test)
print("Random Subspaces classifier",accuracy_score(y_test,y_pred))
```
